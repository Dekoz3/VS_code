```python
# Чтобы скачать и распаковать zip-архив в output папку в Питоне:
import pandas as pd
import numpy as np
from IPython.display import display
#!/usr/bin/env python
import io
import zipfile
import requests  # $ pip install requests

r = requests.get("https://lms.skillfactory.ru/assets/courseware/v1/8a7b3b3f18edf7a3772760221c97139b/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/students_performance.zip")
with r, zipfile.ZipFile(io.BytesIO(r.content)) as archive:
    archive.extractall('Data')
```

```python
df = pd.read_csv('data/students_performance.csv', sep=',')
df.info()
```

```python
# Задание 9.8
# 1 point possible (graded)
# Каков средний балл по чтению у студентов, которые посещали курсы подготовки к экзаменам?
# Округлите ответ до целого числа. 
mask = df["test preparation course"] == "completed"
score = df[mask]["reading score"].mean()
round(score)
```

```python
# Задание 9.9
# 1 point possible (graded)
# Сколько студентов получили 0 баллов по математике?
mask = df["math score"] == 0
df[mask].count()
```

```python
# Задание 9.10
# 1 point possible (graded)
# Проверьте гипотезу: у студентов с оплачиваемым питанием средний балл по математике выше, чем у студентов с льготным питанием.
# В качестве ответа напишите наибольший средний балл по математике среди этих групп студентов.
# Округлите ответ до целого числа.

display(df.head())
mask1 = df["lunch"] == "standard"
mask2 = df["lunch"] != "standard"
display(round(df[mask1]["math score"].mean()))
display(round(df[mask2]["math score"].mean()))
```

```python
# Задание 9.11
# 1 point possible (graded)
# Каков процент студентов, родители которых имеют высшее образование уровня бакалавриата (bachelor's degree)?
# Округлите ответ до целого числа.
# Не указывайте знак %.

percent = df["parental level of education"].value_counts(normalize = True) *100
round(percent)
```

```python
# Задание 9.12
# 1 point possible (graded)
# Насколько медианный балл по письму у студентов в расовой группе А отличается от среднего балла по письму у студентов в расовой группе C?
# Округлите ответ до целого и запишите модуль этого числа.

mask_A = df["race/ethnicity"] == "group A"
mask_C = df["race/ethnicity"] == "group C"

group_A = df[mask_A]["writing score"].median()
group_C = df[mask_C]["writing score"].mean()
print(group_A, group_C)
round(abs(group_A - group_C))
```
