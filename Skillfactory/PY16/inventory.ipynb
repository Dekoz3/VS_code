```python
#инв_259_03042022.csv
import pandas as pd
import numpy as np
import os 

files = os.listdir('inventory')
files.pop(0)
dirname = 'inventory'
files = list(map(lambda name: os.path.join(dirname, name), files))
print(files)
z = len(files)
print(z)
df_names = ['inv'+ str(x) for x in range(14)]
print(df_names)
```

```python


inv1  = pd.read_csv('inventory\инв_259_03042022.csv', decimal=',', thousands=' ', sep=';', encoding='cp1251')
inv1 = inv1.drop(index=0, axis=0)
# display(inv1.info())
list_to_del = ['Кол-во', "Сумма" , "Сумма розн.", "Кол-во книжн. ЕГАИС" , "К." , 'N']
for col in inv1.columns:
    if inv1[col].count() == 0:
        list_to_del.append(col)
inv1 = inv1.drop(columns=list_to_del, axis=1)
display(inv1.head())
for line in range(2, len(inv1.index), 2):
    if inv1.loc[line, 'Кол-во книжн.'] not in ['NaN', np.NaN]:
        inv1.loc[ (line - 1) , 'Кол-во факт.'] = inv1.loc[line, 'Кол-во книжн.']
        inv1 = inv1.drop(index= line, axis=0)
# inv1 = inv1.fillna(0)

files.remove('inventory\инв_259_03042022.csv')

for file in files:
    inv = pd.read_csv(file, decimal=',', thousands=' ', sep=';', encoding='utf8')
    inv = inv.drop(index=0, axis=0)
    list_to_del = ['Кол-во', "Сумма" , "Сумма розн.", "Кол-во книжн. ЕГАИС" , "К.", 'N']
    for col in inv.columns:
        if inv[col].count() == 0:
            list_to_del.append(col)
    inv = inv.drop(columns=list_to_del, axis=1)
    
    for line in range(2, len(inv.index), 2):
        if inv.loc[line, 'Кол-во книжн.'] not in ['NaN', np.NaN]:
            inv.loc[ (line - 1) , 'Кол-во факт.'] = inv.loc[line, 'Кол-во книжн.']
            inv = inv.drop(index= line, axis=0)
    # inv = inv.fillna(0)
    inv1 = pd.concat([inv1 , inv])

inv1 = inv1.dropna(axis = 0, thresh= 5)
display(inv1.head())

writer = pd.ExcelWriter('total_inventory.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
inv1.to_excel(writer, sheet_name='Sheet1', encoding='utf8')
writer.save()
```
